#### import libaries

```python
import pandas as pd
```

#### load csv

```python
og_data = pd.read_csv('world_population.csv')
og_data
```

### checking for missing values

```python
og_data.info()
```

```python
og_data.isna().any()
```

### data wrangling and tidying

#### add column: 'Growth from 1970 to 2020'

```python
og_data['Growth from 1970 to 2020'] = og_data['2020 Population'] / og_data['1970 Population']
og_data
```

#### unpivot table

```python
data_unpivot = pd.melt(og_data, id_vars= ['Rank', 'CCA3', 'Country', 'Capital', 'Continent', 'Area (km²)', 'Density (per km²)', 'Growth Rate', 'World Population Percentage', 'Growth from 1970 to 2020'], value_vars= ['2022 Population', '2020 Population', '2015 Population', '2010 Population', '2000 Population', '1990 Population', '1980 Population', '1970 Population'])
data_unpivot
```

### clean up: headers

```python
data_unpivot['Year'] = data_unpivot.variable.str.replace(' Population', '')
data_unpivot
```

```python
data_unpivot = data_unpivot.rename(columns= {'value': 'Population'})
data_unpivot
```

```python
df = data_unpivot.drop('variable', axis = 1)
df
```

#### sort dataframe

```python
df = df.sort_values(by= ['Rank', 'Year']).reset_index(drop= True)
df
```

### export whole data frame to csv

```python
df.to_csv('00_world_population_unpivoted.csv', sep=';')
```

### slicing continental/regional data frames

#### north america

```python
df_north_america = df[df['Continent'] == 'North America'].sort_values(by= ['Rank', 'Year']).reset_index(drop= True)
df_north_america
```

#### south america

```python
df_south_america = df[df['Continent'] == 'South America'].sort_values(by= ['Rank', 'Year']).reset_index(drop= True)
df_south_america
```

#### europe

```python
df_europe = df[df['Continent'] == 'Europe'].sort_values(by= ['Rank', 'Year']).reset_index(drop= True)
df_europe
```

#### africa

```python
df_africa = df[df['Continent'] == 'Africa'].sort_values(by= ['Rank', 'Year']).reset_index(drop= True)
df_africa
```

#### asia

```python
df_asia = df[df['Continent'] == 'Asia'].sort_values(by= ['Rank', 'Year']).reset_index(drop= True)
df_asia
```

#### ociania

```python
df_oceania = df[df['Continent'] == 'Oceania'].sort_values(by= ['Rank', 'Year']).reset_index(drop= True)
df_oceania
```

#### european union

```python
eu = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden']
df_european_union = df.loc[df.Country.isin(eu),:]
df_european_union
```

##### check if all members are included

```python
if len(eu) == df_european_union.Country.nunique():
    print('all good')
```

### export regional data frames to csv

```python
df_north_america.to_csv('01_north_american_population_unpivoted.csv', sep=';')
df_south_america.to_csv('02_south_american_population_unpivoted.csv', sep=';')
df_europe.to_csv('03_european_population_unpivoted.csv', sep=';')
df_africa.to_csv('04_african_population_unpivoted.csv', sep=';')
df_asia.to_csv('05_asian_population_unpivoted.csv', sep=';')
df_oceania.to_csv('06_oceanian_population_unpivoted.csv', sep=';')
df_european_union.to_csv('07_european_union_population_unpivoted.csv', sep=';')
```
